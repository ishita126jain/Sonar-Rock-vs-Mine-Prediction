```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('/home/ishita/Documents/datasets/sonar data.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
dataset['R'].value_counts()
```

```python
X = dataset.drop(columns='R',axis=1)
```

```python
Y = dataset['R']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test , Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
print(X_train)
print(Y_train)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
```

```python
training_accuracy_score = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_accuracy_score)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
testing_accuracy_score = accuracy_score(X_test_prediction, Y_test)
```

```python
print(testing_accuracy_score)
```
